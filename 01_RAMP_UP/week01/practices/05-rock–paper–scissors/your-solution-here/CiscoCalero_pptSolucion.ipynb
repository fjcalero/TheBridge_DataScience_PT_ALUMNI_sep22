```python
# Importa las funciones del script
import utils

# Asigna a una lista las 3 posibles opciones. 
opciones_juego = ["Piedra", "Papel", "Tijera"]
combinaciones_ganadoras = [("Piedra", "Tijera"), ("Tijera", "Papel"), ("Papel", "Piedra")]

#Numero de partidas impar
num_partidas = 5

# Numero de partidas que debe ganar basada en el numero de partidas
partidas_ganadas = 5 // 2 + 1

# Crea dos variables que acumulen las partidas ganadas de cada participante
ganaHumano = 0
ganaMaquina = 0

#Crea un bucle que itere mientras ningún jugador alcance el mínimo de victorias
# necesarias para ganar. Dentro del bucle resuelve la jugada de la
# máquina y pregunta la del jugador. Comparalas y actualiza el valor 
# de las variables que acumulen las partidas ganadas de cada participante.

contador = 0
todas_jugadas_maquina = []
todas_jugadas_humano = []

while contador < num_partidas:
    jugada_maquina = utils.jugadaMaquina (opciones_juego)  
    jugada_humano = utils.jugadaHumano (opciones_juego)
    resultado_jugada = utils.combate (jugada_maquina, jugada_humano, combinaciones_ganadoras)
    if resultado_jugada == 1:
        ganaMaquina +=1
    elif resultado_jugada == 2:
        ganaHumano += 1
    todas_jugadas_maquina.append (jugada_maquina)
    todas_jugadas_humano.append (jugada_humano)
    if ganaMaquina >= partidas_ganadas or ganaHumano >= partidas_ganadas:
        break
    contador += 1

utils.estadoPartida (todas_jugadas_maquina, todas_jugadas_humano)
print("******"*5)
print(f"Jugadas ganadas máquina: {ganaMaquina}")
print(f"Jugadas que tu has ganado: {ganaHumano}")
print("******"*5)

# Anuncia por consola el ganador del juego en función de quién tiene más victorias aculumadas
if ganaMaquina > ganaHumano:
    print("Gana la máquina")
elif ganaHumano > ganaMaquina:
    print("Tu has ganado")
else:
    print("Habeis quedado empate!")
```
